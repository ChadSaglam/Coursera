### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here

# Check for missing values in each column
print(df.isnull().sum())

# Drop rows with missing values in key columns (e.g., 'JobSat', 'YearsCodePro')
df_cleaned = df.dropna(subset=['JobSat', 'YearsCodePro'])

# Alternatively, fill missing values with a placeholder (e.g., 'Unknown')
df_filled = df.fillna('Unknown')
```

```python
# Check for missing values in each column
print(df.isnull().sum().sort_values(ascending=False))
```

```python
# Calculate the percentage of missing values in each column (%)
missing_percentage = df.isnull().sum() / len(df) * 100

# Select columns with more than 30% missing data
columns_to_drop = missing_percentage[missing_percentage > 56].index

# Drop these columns from the DataFrame
df_cleaned = df.drop(columns=columns_to_drop)

# Print the dropped columns and the shape of the new DataFrame
print(f"Dropped columns: {columns_to_drop.tolist()}")
print(f"New DataFrame shape: {df_cleaned.shape}")
```

```python
print(missing_percentage.sort_values(ascending=False))
```

```python
# Check for missing values in each column
print(df_cleaned.isnull().sum().sort_values(ascending=False))
```

```python
print(df_filled)
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
## Write your code here

# Calculate value counts for key columns
print(df['Employment'].value_counts())
print(df['JobSat'].value_counts())
print(df['YearsCodePro'].value_counts())
```

```python
# Calculate value counts for key columns
print(df['Employment'].isnull().sum())
print('---------------------------------------')
print(df['Employment'].value_counts().sort_values(ascending=False))
```

```python
# Create a new column for simplified employment status
df['Employment_Simplified'] = df['Employment'].apply(lambda x: x.split(';')[0] if ';' in x else x)

# Check the new distribution
print(df['Employment_Simplified'].value_counts())
```

```python
# Define a mapping for broader categories
employment_mapping = {
    'Employed, full-time': 'Employed',
    'Employed, part-time': 'Employed',
    'Independent contractor, freelancer, or self-employed': 'Self-Employed',
    'Student, full-time': 'Student',
    'Student, part-time': 'Student',
    'Not employed, but looking for work': 'Not Employed',
    'Not employed, and not looking for work': 'Not Employed',
    'Retired': 'Other',
    'I prefer not to say': 'Other'
}

# Apply the mapping to create a new column
df['Employment_Grouped'] = df['Employment_Simplified'].map(employment_mapping)

# Check the new distribution
print(df['Employment_Grouped'].value_counts())
```

```python
print(df['JobSat'].isnull().sum())
print('---------------------------------------')
print(df['JobSat'].value_counts().sort_values(ascending=False))
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here

# Create a pie chart for JobSat
df['JobSat'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Job Satisfaction Distribution')
plt.ylabel('')
plt.show()
```

```python
# Alternatively, create a KDE plot
sns.kdeplot(df['JobSat'].dropna(), shade=True)
plt.title('Job Satisfaction Distribution (KDE)')
plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here

# Compare programming languages worked with and wanted to work with
languages_worked = df['LanguageHaveWorkedWith'].str.get_dummies(sep=';').sum()
languages_wanted = df['LanguageWantToWorkWith'].str.get_dummies(sep=';').sum()

# Create a grouped bar chart
pd.DataFrame({'Worked With': languages_worked, 'Want to Work With': languages_wanted}).plot(kind='bar')
plt.title('Programming Languages: Worked With vs. Want to Work With')
plt.xlabel('Programming Language')
plt.ylabel('Count')
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here

# Visualize RemoteWork distribution by region
sns.countplot(data=df, x='RemoteWork', hue='Country')
plt.title('Remote Work Distribution by Region')
plt.xlabel('Remote Work Preference')
plt.ylabel('Count')
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here

# Calculate Pearson correlation between JobSat and YearsCodePro
correlation = df['JobSat'].astype('category').cat.codes.corr(df['YearsCodePro'].astype('category').cat.codes)
print(f'Correlation between Job Satisfaction and Years of Professional Coding: {correlation:.2f}')
```

```python
sns.scatterplot(x=df['YearsCodePro'], y=df['JobSat'])
plt.title('Job Satisfaction vs. Years of Professional Coding')
plt.xlabel('Years of Professional Coding')
plt.ylabel('Job Satisfaction')
plt.show()
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here

# Create a cross-tabulation of Employment and EdLevel
cross_tab = pd.crosstab(df['Employment_Simplified'], df['EdLevel'])

# Visualize the cross-tabulation with a stacked bar plot
cross_tab.plot(kind='bar', stacked=True)
plt.title('Employment Status vs. Education Level')
plt.xlabel('Employment Status')
plt.ylabel('Count')
plt.show()
```
